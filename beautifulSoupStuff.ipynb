```python
from bs4 import BeautifulSoup
import requests
```

```python
URL = "https://coinmarketcap.com"
```

```python
html = requests.get(URL)
doc = BeautifulSoup(html.text,"html.parser")
```

```python
tbody = doc.tbody
trs = tbody.contents
prices = {}
for tr in trs:
    for td in tr.contents:
        print(td)
        print("")
```
